```python
# Начальные условия
t = 0
N = 500
x = 1
k = 0.01
dt = 0.01
```

```python
# Списки для хранения точек x и t
x_a = []
t_a = []
```

```python
x_a.append(x)
t_a.append(t)
```

```python
# Правая часть уравнения
def f(x):
  return k * x * (N - x)
```

```python
# Метод Эйлера
while x <= 0.95 * N:
  x = x + dt * f(x)
  t = t + dt
  x_a.append(x)
  t_a.append(t)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t_a, x_a, linestyle='-', color='b', label="Количество клиентов")
plt.title("Количество клиентов от времени", fontsize=14)
plt.xlabel("Время t", fontsize=12)
plt.ylabel("Количество клиентов", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```

```python
print(f"В момент времени {t_a[-1]:.2f} число покупателей стало равно 0.95*N")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Правые части уравнений
def f(x, y):
  return a1*x + x*x*y + x*x*x*y + x*x*x*x*y

def g(x, y):
  return a2*y + y*y*x + y*y*y*x + y*y*y*y*x

# Коэффециенты
a2 = -2
a1 = -1

# Вычисление производных
def equations(x, y):
    dxdt = f(x, y)
    dydt = g(x, y)
    return dxdt, dydt

# Метод Эйлера
def euler_method(t0, t_end, dt, initial_conditions):
    t_values = np.arange(t0, t_end, dt) # Значения t
    num_steps = len(t_values)

    x = np.zeros(num_steps)
    y = np.zeros(num_steps)

    x[0], y[0] = initial_conditions # Начальные точки одной траектории

    # Метод Эйлера
    for i in range(num_steps - 1):
        dxdt, dydt = equations(x[i], y[i])

        x[i + 1] = x[i] + dxdt * dt
        y[i + 1] = y[i] + dydt * dt

    return t_values, x, y

# Начальное и конечное t
t0 = 0
t_end = 20
dt = 0.01

# Начальные точки, с которых начинаются траектории
initial_points = [
    [0.5, 0.5],
    [-0.5, -0.5],
    [-0.75, 0.5],
    [0.1, 0.75],
    [0.75, -0.5],
    [1, 0.5]
]

plt.figure(figsize=(8, 6))

for initial_conditions in initial_points:
    t_values, x, y = euler_method(t0, t_end, dt, initial_conditions)

    plt.plot(x, y, color='b')

    # Добавление стрелок вдоль траектории
    skip = 15 # частота стрелок
    for i in range(0, len(x) - skip - 1850, skip):
      plt.arrow(x[i], y[i], x[i + skip] - x[i], y[i + skip] - y[i],
                  head_width=0.015, head_length=0.05, fc='r', ec='r', length_includes_head=True)

    plt.xlabel("Ось x")
    plt.ylabel("Ось y")
    plt.title("Фазовый портрет системы уравнений")
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import random as rnd
```

```python
# Модель игральной кости - случайное целое число из диапазона [1, 6]
def dice():
  return int(rnd.random() * 6 + 1)
```

```python
# Один эксперинт - 500 подбрасываний и подсчет матожидания
def experiment():
  sum = 0
  for i in range(500):
    sum += dice()
  ev = sum / 500
  return ev
```

```python
experiments = []
```

```python
# Десять экспериментов
for i in range(10):
  exp = experiment()
  print(f"Испытание {i+1}: матожидание 500 бросков: {exp}")
  experiments.append(exp)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(list(range(1, len(experiments) + 1)), experiments, marker='o', linestyle='-', color='b', label="Матожидание")
plt.title("Матожидание для 500 бросков в каждом испытании", fontsize=14)
plt.xlabel("Номер испытания", fontsize=12)
plt.ylabel("Матожидание", fontsize=12)
plt.xticks(list(range(1, len(experiments) + 1)))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```

```python
# Генерация точки - по оси X случайное целое число из отрезка [0, a], по оси Y - [0, b]
def dot(a, b):
  return rnd.random() * a, rnd.random() * b
```

```python
# Количество точек в каждом испытании
num_dots = [30, 60, 100, 1000, 10000]
```

```python
squares = []
for num_dot in num_dots:
  in_circle = 0
  for i in range(num_dot):
    x, y = dot(2, 2)
    if (x-1) ** 2 + (y-1) ** 2 <= 1: # Попала ли точка в круг с центром (1, 1) и радиусом 1
      in_circle += 1
  sq = in_circle / num_dot * 4 # Домножаем на площадь квадрата - 4
  squares.append(sq)
  print(f"Для {num_dot} точек площадь круга = {sq}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(num_dots, squares, marker='o', linestyle='-', color='b', label="Площадь круга")
plt.title("Площадь круга на количество точек", fontsize=14)
plt.xlabel("Количество точек", fontsize=12)
plt.ylabel("Площадь", fontsize=12)
plt.xscale('log')
plt.xticks(num_dots)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```

```python
# Матожидание и отклонение хронометра
ev = 0
sigma = 1
```

```python
# Нормальное распределение
def normal(ev, sigma):
  sum = 0
  for i in range(12):
    sum += rnd.random()
  return ev + sigma * (sum - 6)
```

```python
# Погрешность одного хронометра за неделю
def chronometer_per_week():
  week = 0
  for i in range(7):
    week += normal(ev, sigma)
  return week
```

```python
# Усредненная погрешность трех хронометров за неделю
def mean_error_per_week():
  three_chronometers = 0
  for i in range(3):
    three_chronometers += chronometer_per_week()
  return three_chronometers / 3
```

```python
k_errors = 0
for i in range(1000):
  error = mean_error_per_week()
  if error > 2: # Если величина ошибки превышает две минуты
    k_errors += 1
print(f"На основании 1000 испытаний вероятность того, что погрешность больше двух минут = {k_errors / 1000}")
```

```python
import math
```

```python
# Начальные условия
z = 0
H = 10
lm = 0.5
p = 0.3
```

```python
# Экспоненциальное распределение
def exponential(lm):
  return -1/lm * math.log(rnd.random())
```

```python
success = 0

for i in range(1000):
  z = 0
  while True:
    z += exponential(lm)
    if z >= H: # Если частица достигла точки H - успех
      success += 1
      break
    if rnd.random() < p: # Частица прекращает движение
      break

print(f"Вероятность достигнуть точки H на 1000 испытаний = {success / 1000}")
```

```python
import numpy as np

# Параметры системы
lambda_rate = 1.0  # Интенсивность поступления заявок
mu_rate = 1.2      # Интенсивность обслуживания
queue_capacity = 4 # Вместимость очереди
max_capacity = 5   # Максимальное число заявок в системе (очередь + обслуживание)
simulation_time = 1000

# Инициализация переменных
np.random.seed(42)
current_time = 0
next_arrival = np.random.exponential(1 / lambda_rate)
next_departure = float('inf')

queue = 0
served = 0
rejected = 0
total_in_system = 0

# Массив для хранения состояния системы
states = []

# Основной цикл моделирования
while current_time < simulation_time:
    if next_arrival < next_departure:
        # Обработка прибытия заявки
        current_time = next_arrival
        if queue + 1 < max_capacity:
            if queue < queue_capacity:
                queue += 1
                if queue == 1:  # Если заявка начинает обслуживание
                    next_departure = current_time + np.random.exponential(1 / mu_rate)
            else:
                rejected += 1
        else:
            rejected += 1
        next_arrival = current_time + np.random.exponential(1 / lambda_rate)
    else:
        # Обработка завершения обслуживания
        current_time = next_departure
        served += 1
        queue -= 1
        if queue > 0:
            next_departure = current_time + np.random.exponential(1 / mu_rate)
        else:
            next_departure = float('inf')

    total_in_system += queue
    states.append(queue)

# Оценка характеристик системы
u = rejected / (served + rejected)  # Вероятность отказа
n = total_in_system / simulation_time  # Среднее число заявок в системе

print(f"Вероятность отказа (u): {u:.4f}")
print(f"Среднее число заявок в системе (n): {n:.4f}")
```
