```python
import csv

table = []
```

```python
with open('temp/task19.csv', 'r') as file:
    reader = csv.DictReader(file, delimiter=';')
    for row in reader:
        table.append(row)
table[:5:]
```

```python
sum_dist = 0

weight_osinki = 0
count_osinki = 0

gasoline = 0

weight_berezki = 0
count_berezki = 0

for row in table:
    
    # 1. На какое суммарное расстояние были произведены перевозки с 7 по 9 октября?
    if '7 октября' <= row['Дата'] <= '9 октября':
        sum_dist += int(row['Расстояние'])

    # 2. Какова средняя масса груза при автоперевозках, осуществлённых из города Осинки? 
    if row['Пункт отправления'] == 'Осинки':
        weight_osinki += int(row['Масса груза'])
        count_osinki += 1
        
    # 3. Какой суммарный расход бензина был при осуществлении перевозок с 1 по 3 октября? 
    if '1 октября' <= row['Дата'] <= '3 октября':
        gasoline += int(row['Расход бензина'])
        
    # 4. Какова средняя масса груза при автоперевозках, осуществлённых в город Березки? 
    if row['Пункт назначения'] == 'Березки':
        weight_berezki += int(row['Масса груза'])
        count_berezki += 1
```

```python
print(f"1. Суммарная масса груза перевозок с 7 по 9 октября: {sum_dist}")
print(f"2. Cредняя масса груза при автоперевозках, осуществлённых из города Осинки: {weight_osinki/count_osinki}")
print(f"3. Суммарный расход бензина при перевозках с 1 по 3 октября: {gasoline}")
print(f"4. Средняя масса груза при автоперевозках, осуществлённых в город Березки: {weight_berezki/count_berezki}")
```
